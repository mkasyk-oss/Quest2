```python
import networkx as nx
import matplotlib.pyplot as plt

# граф
G = nx.DiGraph()

# Вузл
G.add_node("Програміст", type='person')
G.add_node("Комп'ютер", type='object')
G.add_node("Програма", type='object')
G.add_node("Сів за комп'ютер", type='action')
G.add_node("Налагодив програму", type='action')

# звязки
G.add_edge("Програміст", "Сів за комп'ютер")
G.add_edge("Сів за комп'ютер", "Комп'ютер")

G.add_edge("Програміст", "Налагодив програму")
G.add_edge("Налагодив програму", "Програма")


pos = {
    "Програміст": (0,1),
    "Сів за комп'ютер": (-1,0),
    "Комп'ютер": (-1,-1),
    "Налагодив програму": (1,0),
    "Програма": (1,-1)
}

nx.draw(G, pos, with_labels=True, arrows=True, node_size=3000, node_color='lightblue', font_size=10)
plt.title("Семантична сітка Програміст сів за комп’ютер та налагодив програму")
plt.show()
```
