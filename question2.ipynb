```python
import numpy as np
import matplotlib.pyplot as plt

# Функція
def hot_tea_membership(t):
    if t <= 50:
        return 0
    elif 50 < t < 70:
        return (t - 50) / 20
    elif 70 <= t <= 90:
        return 1
    elif 90 < t < 100:
        return (100 - t) / 10
    else:
        return 0


temperatures = np.linspace(30, 110, 500)
membership_values = [hot_tea_membership(t) for t in temperatures]

#графік
plt.plot(temperatures, membership_values, label='Гарячий чай')
plt.xlabel('Температура °C')
plt.ylabel('Належність μ')
plt.title('"Гарячий чай"')
plt.grid(True)
plt.legend()
plt.show()
```
