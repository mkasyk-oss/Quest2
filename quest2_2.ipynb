```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

# фрейми
G.add_node("Птах", type='class')

G.add_node("Соловей", type='subclass', attributes={"Літає":"Так", "Звук":"Щебет", "Їжа":"Комахи"})
G.add_node("Пінгвін", type='subclass', attributes={"Літає":"Ні", "Звук":"Крик", "Їжа":"Риба"})
G.add_node("Сова", type='subclass', attributes={"Літає":"Так", "Звук":"Угук", "Їжа":"Дрібні ссавці"})

# Зєдннання
G.add_edge("Птах", "Соловей")
G.add_edge("Птах", "Пінгвін")
G.add_edge("Птах", "Сова")

# Розташування
pos = {
    "Птах": (0,2),
    "Соловей": (-2,0),
    "Пінгвін": (0,0),
    "Сова": (2,0)
}

# вузли
nx.draw_networkx_nodes(G, pos, node_size=3000, node_color='lightgreen')
nx.draw_networkx_labels(G, pos, font_size=10)

# ребра
nx.draw_networkx_edges(G, pos, arrows=True)

#атрибути
for node in G.nodes(data=True):
    if 'attributes' in node[1]:
        attrs = "\n".join(f"{k}: {v}" for k,v in node[1]['attributes'].items())
        x, y = pos[node[0]]
        plt.text(x, y-0.4, attrs, fontsize=9, ha='center')

plt.title("Мережа фреймів: Птахи")
plt.axis('off')
plt.show()
```
